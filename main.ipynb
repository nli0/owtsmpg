## Arizona

Through the passage of Proposition 106 in 2000, the voters of Arizona transferred the responsibility of drawing congressional boundaries from the state legislature to the newly established [Independent Redistricting Commission](https://irc.az.gov/) (IRC), a bipartisan commission of members of the public. Had Proposition 106 not passed, redistricting in Arizona would have remained in the hands of the Republican-controlled legislature.

Following the 2010 Census, the IRC constructed a new map which was generally seen as [slightly favorable](https://www.washingtonpost.com/blogs/the-fix/post/redistricting-draft-map-in-arizona-favors-democrats/2011/10/04/gIQAI33YLL_blog.html) to the Democratic Party. In 2012, for example, the Democratic Party secured a majority of the Arizona's congressional seats despite losing the statewide popular vote by [more than 8 points](https://en.wikipedia.org/wiki/2012_United_States_House_of_Representatives_elections_in_Arizona). Over the past decade, the Democratic Party has consistently enjoyed a slight advantage in partisan bias in spite of the IRC's nonpartisan nature. In the 2016 House of Representatives elections, the partisan bias actually swung towards the Republican Party; however, this can be mostly attributed to the lack of a Republican general election challenger to the Democratic incumbent in Arizona's 3rd Congressional District.

```python
plot.draw('AZ', (10, 10))
```

## California

Following in the footsteps of Arizona, the voters of California passed Proposition 20 in 2010, transferring the responsibility of redistricting to the [California Citizens Redistricting Commission](https://wedrawthelines.ca.gov/index/) (CCRC), a similar independent bipartisan commission to Arizona's IRC. While the CCRC had been established prior to the 2010 election, it was only responsible for drawing the boundaries of the California State Assembly, Senate, and Board of Equalization.

After 2010, California has consistently featured highly competitive congressional elections with low partisan bias. In contrast to Arizona, however, California is a solidly blue state that has far more Democrats than Republicans, so it may also be useful to measure the partisan bias at a actual vote margin closer to the partisan composition of the state (roughly D+30).

```python
plot.draw('CA', (10, 10))
```

## Maryland

Like California, Maryland is a solidly blue state controlled by the Democratic Party at most levels of government. However, the responsibility of drawing congressional district boundaries primarily lies in the state legislature, which has repeatedly used its power to draw maps favorable for the party in control, best exemplified by high partisan bias towards the Democratic Party.

Following the 2010 Census, the Democratic Party could have plausibly drawn all eight districts to favor their party. However, many Democratic incumbents wanted to keep certain communities within their districts, resulting in the current [map](https://en.wikipedia.org/wiki/Maryland%27s_congressional_districts#/media/File:Maryland_Congressional_Districts,_113th_Congress.tif), which has yielded 7 Democratic districts and 1 Republican district for the 2012 - 2018 elections.

```python
plot.draw('MD', (10, 10))
```

## Wisconsin
In the 2010 midterm elections, the Republican Party [swept control](https://ballotpedia.org/Wisconsin_2010_legislative_election_results) of both the Wisconsin State House and State Senate, gaining control of redistricting for Wisconsin's seats in the House of Representatives. Republicans drew 5 out of 8 districts for themselves, allowing them to control the majority of Wisconsin's congressional delegation throughout the 2010s, in spite of the Democratic Party winning the overall popular vote in 2012, 2016, and 2018. Furthermore, Wisconsin is a difficult state for the Democratic Party to win legislative elections in, as many of their voters are "packed" within Milwaukee and Dane County.
